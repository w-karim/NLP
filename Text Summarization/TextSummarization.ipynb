```python
import string
import pandas as pd
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
```

```python
parag = """So, keep working. Keep striving. 
        Never give up. Fall down seven times, get up eight. 
        Ease is a greater threat to progress than hardship. 
        Ease is a greater threat to progress than hardship. 
        So, keep moving, keep growing, keep learning. 
        See you at work."""
```

```python
parag_tokenize = sent_tokenize(parag)
parag_tokenize
```

```python
import string

cleantokens =[]

stop_words = set(stopwords.words('english'))
tokens = word_tokenize(parag)
tokens = list(filter(lambda token: token not in string.punctuation, tokens))

for word in tokens: 
    if word not in stop_words:
        cleantokens.append(word.lower())
cleantokens
```

```python
word_frequencies = {}
for word in cleantokens:
    if word in word_frequencies:
        word_frequencies[word] += 1
    else:
        word_frequencies[word] = 1
all_values = word_frequencies.values()
maxi = max(all_values)
word_frequencies
```

```python
df = pd.DataFrame(columns = ['Word','Frequency','Weighted Frequency'])
for i in range(len(word_frequencies)):
    df.loc[i] = [list(word_frequencies.keys())[i], list(word_frequencies.values())[i], list(word_frequencies.values())[i]/maxi]
df
```

scoring the sentences

```python
parag_tokenize
```

```python
sentence_scores = {}
for sent in parag_tokenize:
    for word in word_tokenize(sent.lower()):
        if word in word_frequencies.keys():
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = word_frequencies[word]
                else:
                    sentence_scores[sent] += word_frequencies[word]
```

```python
df2 = pd.DataFrame(columns = ['Sentence','Score'])
for i in range(len(sentence_scores)):
    df2.loc[i] = [list(sentence_scores.keys())[i], list(sentence_scores.values())[i]]
df2
```

Summarize

```python
import heapq
```

```python
summary_sentences = heapq.nlargest(2, sentence_scores, key=sentence_scores.get )
summary = ' '.join(summary_sentences)
print(summary)
```

```python
import bs4 as bs
import urllib.request
import string
import pandas as pd
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import heapq
import re
```

```python
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Natural_language_processing').read()
soup = bs.BeautifulSoup(source,'lxml')
paragraphs = soup.find_all('p')

article_text = ""

for p in paragraphs:
    article_text += p.text

article_text = re.sub(r'\[[0-9]*\]', ' ', article_text)
article_text = re.sub(r'\s+', ' ', article_text)
formatted_article_text = re.sub('[^a-zA-Z]', ' ', article_text )
formatted_article_text = re.sub(r'\s+', ' ', formatted_article_text)
print("Long string with newlines: \n" + article_text)
```
